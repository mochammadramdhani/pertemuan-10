```python
# Import Library yang akan digunakan
# %matplotlib inline  # Perintah ini digunakan di Jupyter Notebook untuk menampilkan plot langsung di dalam notebook

import matplotlib.pyplot as plt  # Mengimpor matplotlib untuk membuat grafik/plot dan menggunakan alias `plt`
import numpy as np  # Mengimpor numpy untuk operasi numerik pada array dan matriks, menggunakan alias `np`
import pandas as pd  # Mengimpor pandas untuk manipulasi data tabular, menggunakan alias `pd`
from sklearn.cluster import KMeans  # Mengimpor algoritma K-Means dari library scikit-learn untuk clustering
```

```python
# Menyiapkan data dan memanggil dataset
dataset = pd.read_csv('konsumenn.csv')
# Membaca file CSV dengan nama 'konsumen.csv'
# `pd.read_csv` adalah fungsi dari pandas untuk membaca file CSV dan mengubahnya menjadi DataFrame.

dataset.keys()
# Menampilkan daftar kolom (keys) dalam dataset untuk memberikan gambaran tentang struktur data.
```

```python
dataku = pd.DataFrame(dataset)
# Membuat DataFrame pandas baru dengan nama `dataku` dari data yang telah dimuat ke variabel `dataset`.
# DataFrame adalah struktur data berbentuk tabel 2D yang sangat fleksibel untuk manipulasi data.

dataku.head()
# Menampilkan 5 baris pertama dari DataFrame `dataku` secara default.
# Fungsi ini berguna untuk memeriksa dan melihat cuplikan awal data.
```

```python
# Konversi ke data Array
X = np.asarray(dataset)
# Mengonversi objek `dataset` (misalnya DataFrame atau list) menjadi array NumPy.
# `np.asarray()` adalah fungsi NumPy yang mengubah input ke dalam bentuk array.

print(X)
# Mencetak array `X` untuk memastikan bahwa data telah berhasil dikonversi.
```

```python
# Menampilkan data dalam bentuk scatter plot
plt.scatter(X[:, 0], X[:, 1], label='True Position')
# Membuat scatter plot untuk data pada array `X`.
# - `X[:, 0]`: Kolom pertama dari array (misalnya, "Gaji").
# - `X[:, 1]`: Kolom kedua dari array (misalnya, "Pengeluaran").
# - `label='True Position'`: Label untuk legenda pada plot.

plt.xlabel("Gaji")
# Menambahkan label pada sumbu x dengan teks "Gaji".

plt.ylabel("Pengeluaran")
# Menambahkan label pada sumbu y dengan teks "Pengeluaran".

plt.title("Grafik Penyebaran Data Konsumen")
# Menambahkan judul pada grafik dengan teks "Grafik Penyebaran Data Konsumen".

plt.show()
# Menampilkan plot yang telah dibuat.
```

```python
# Mengaktifkan K-Means dengan jumlah K=2
kmeans = KMeans(n_clusters=2)
# Membuat model K-Means dengan jumlah klaster sebanyak 2.
# - `n_clusters=2`: Menentukan jumlah klaster yang akan dibentuk oleh model.

kmeans.fit(X)
# Melatih model K-Means dengan data yang terdapat pada array `X`.
# `fit()` menghitung posisi centroid dan menentukan data mana yang termasuk dalam setiap klaster.
```

```python
# Menampilkan nilai Centroid yang digenerate oleh algoritma
print(kmeans.cluster_centers_)
# Mencetak nilai centroid yang dihasilkan oleh algoritma K-Means.
# `kmeans.cluster_centers_` berisi koordinat centroid untuk setiap klaster.
```

```python
# Plot Data Point
# Memvisualisasikan Hasil Klasterisasi Data Konsumen
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
# Membuat scatter plot berdasarkan data `X`.
# - `X[:, 0]`: Data pada kolom pertama (misalnya, Gaji).
# - `X[:, 1]`: Data pada kolom kedua (misalnya, Pengeluaran).
# - `c=kmeans.labels_`: Warna data berdasarkan label klaster hasil K-Means.
# - `cmap='rainbow'`: Menggunakan colormap 'rainbow' untuk memberikan warna berbeda pada setiap klaster.

plt.xlabel("Gaji")
# Menambahkan label pada sumbu x dengan teks "Gaji".

plt.ylabel("Pengeluaran")
# Menambahkan label pada sumbu y dengan teks "Pengeluaran".

plt.title("Grafik Hasil Klasterisasi Data Gaji dan Pengeluaran Konsumen")
# Menambahkan judul pada grafik untuk menjelaskan isi plot.

plt.show()
# Menampilkan grafik yang telah dibuat.
```

```python
# Plot Data Point
# Memvisualisasikan hasil klasterisasi dengan centroid dari masing-masing klaster
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
# Membuat scatter plot untuk data `X` dengan warna berdasarkan label klaster.
# - `X[:, 0]`: Data pada kolom pertama (misalnya, Gaji).
# - `X[:, 1]`: Data pada kolom kedua (misalnya, Pengeluaran).
# - `c=kmeans.labels_`: Warna titik data berdasarkan klaster hasil K-Means.
# - `cmap='rainbow'`: Colormap digunakan untuk membedakan warna tiap klaster.

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black')
# Menambahkan scatter plot untuk menampilkan posisi centroid klaster.
# - `kmeans.cluster_centers_[:, 0]`: Koordinat x dari centroid klaster.
# - `kmeans.cluster_centers_[:, 1]`: Koordinat y dari centroid klaster.
# - `color='black'`: Warna hitam untuk membedakan centroid dari titik data.

plt.xlabel("Gaji")
# Menambahkan label sumbu x dengan teks "Gaji".

plt.ylabel("Pengeluaran")
# Menambahkan label sumbu y dengan teks "Pengeluaran".

plt.title("Grafik Hasil Klasterisasi Data Gaji dan Pengeluaran Konsumen")
# Menambahkan judul untuk grafik agar mudah dipahami.

plt.show()
# Menampilkan grafik yang telah dibuat.
```
